```python
import  json
import pdfplumber
```

```python
pdf_path = r"C:\Users\Dell\Downloads\ML\RAG\data\pdf\Time Series Analysis -Unit 5.pdf"
chunks = []

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        chunks.append({
            "id": f"chunks_{i+1}",
            "title": "Unit 5 Time Series",
            "content": text,
            "page_number": i+1
        })
```

```python
with open("Time_series.json","w", encoding = "utf-8") as f:
    json.dump(chunks, f, ensure_ascii = False, indent = 2)
```
